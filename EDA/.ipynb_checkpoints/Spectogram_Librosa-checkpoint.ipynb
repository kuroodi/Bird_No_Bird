# Use Python/LibROSA To Convert WAV file to Spectrogram

```python
import os
import matplotlib
import pylab
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.preprocessing import StandardScaler
%matplotlib inline
```

### Paths to WAV files

```python

path_to_wav_file_bird = "../../../Desktop/Bird_No_Bird/BirdVox/wav/001b8d8a-4247-4c46-a11b-b868fa3edb0a.wav"
path_to_wav_file_nobird = "../../../Desktop/Bird_No_Bird/BirdVox/wav/00106202-f61e-467d-a80f-070d90421952.wav"


path_to_img_file = "./test_img.jpg"
```

### Convert WAV file to Spectrogram

```python
y, sr = librosa.load(path_to_wav_file_bird)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
plt.figure(figsize=(12,6))
S_power_to_db = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_power_to_db, x_axis='time', y_axis='mel')
# Put a descriptive title on the plot
plt.title('mel spectrogram (power to db)')
# draw a color bar
plt.colorbar(format='%+02.0f dB');
plt.savefig(path_to_img_file)
```

### Save Spectrogram as Grey-Scale PNG (no labels)

```python
path_to_img_file = "./test_img_grey_scale.png"

y, sr = librosa.load(path_to_wav_file_bird)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
plt.figure(figsize=(2.56,2.56))
S_power_to_db = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(data=S_power_to_db, cmap='gray_r')

#save PNG
plt.savefig(path_to_img_file)
```
